```python
# If
if 5 < 10:
    print('Verdade')
```

```python
#If / Elif / Else
num1 = 0
num2 = 10
if num1 < num2:
    print('Num1 é menor que Num2')
elif num1 > num2:
    print('Num1 é maior que Num2')
else:
    print('Num1 é igual a Num2')
```

```python
# Condicional com operadores lógicos
num1 = 10
num2 = 5
num3 = 2
if (num1 > num2) and (num1 > num3):
    print(f"{num1} é maior que {num2} e {num3}")
else:
  print(f"{num1} não é o maior número.")
```

```python
#Exemplo com and, or e not

idade = 18
tem_autorizacao = True
tem_ingresso = True

if (idade >= 18 and tem_ingresso) or (idade < 18 and tem_autorizacao):
    print("Você pode entrar.")
else:
    print("Você não pode entrar.")
```
