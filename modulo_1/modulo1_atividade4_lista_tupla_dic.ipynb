**Listas**

```python
# Listas - Usa colchetes [], Multável, Index

frutas = ["Maçã", "Banana", "Laranja"] # Index: 0,1,2
print(frutas)  # ['Maçã', 'Banana', 'Laranja']

# Acessar elementos
print(frutas[0])
print(frutas[-1]) # Último item

# Modificando elementos
frutas[1] = "Uva" # Muda o index 1, de Banana para Uva
print(frutas)

# Adicionando elementos
frutas.append("Manga") # Append, adiciona no final
print(frutas)

frutas.insert(0,"Melancia") # Insert, adiciona a partir do index
print(frutas)

frutas.remove("Maçã") # Remove pelo valor
print(frutas)
frutas.pop(0) # Remove pelo index
print(frutas)

print(len(frutas)) # Tamanho da lista



```

**Tupla**

```python
# Tuplas são imutáveis, são definidas com parênteses ( )

cores = ("Vermelho", "Verde", "Azul")
print(cores)

print(cores[1]) # Imprime o index 1

#     cores[1] = "Amarelo"    - Caso tente imprimir, dá erro, pois é imultável.

# Convertendo tupla em lista, caso precise mudar
cores_lista = list(cores) # Criei uma var e atribui os itens da tupla
cores_lista.append("Amarelo")

cores = tuple(cores_lista) # Transforma a lista em tupla
print(cores) # Imprime a tupla com o item "Amarelo"

print(len(cores))
```

**Dicionários**

```python
# Dicionários  são estruturas de dados que armazenam pares de chave-valor {chave, valor}
pessoa = { # pessoa - dicinário
    "nome": "Alice",
    "idade": 25,
    "cidade": "São Paulo"
}
print(pessoa)

# Acessando valores de um dicinário
print(pessoa["nome"])
print(pessoa["idade"])

# Modificando valores
pessoa["idade"] = 26
print(pessoa)

# Adicionando valores
print(pessoa.get("cidade"))
print(pessoa.get("email", "Chave do email não encontrada."))

# Adicionando chave e valores
pessoa["profissao"] = "Desenvolvedor"
print(pessoa)

# Deletando itens
del pessoa ["idade"]
print(pessoa)

# Percorrendo um dicionário
# Apenas chaves
for chave in pessoa:
    print(chave) # -> nome, cidade, profissão
# Apenas valores
for valor in pessoa.values():
    print(valor) # -> Alice, São Paulo, Desenvlvedor
# Chave e valores
for chave, valor in pessoa.items():
    print(f"{chave}: {valor}") # -> "nome: Alice..."

# Verificando se uma chave existe
if "nome" in pessoa:
    print("A chave 'nome' existe no dicionário!")
```

```python
# Dicionário dentro de dicionário

aluno = {
    "nome": "Lucas",
    "notas": {"matemática": 8.5, "história": 7.0}
}
print(aluno["notas"]["matemática"]) # -> 8.5

# Convertando listas e dicionários

lista_tuplas = [("nome", "Carlos"), ("idade", 30)]
dicionario = dict(lista_tuplas)
print(dicionario) # {'nome': 'Carlos', 'idade': 30}

```
