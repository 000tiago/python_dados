**Estutura For, While**

```python
# For é usado para iterar sobre uma sequência, como lista, tupla, string 'elemento/sequencia'

nomes = ["Ana", "Bruno", "Carlos"]
for nome in nomes:
    print(nome)
```

```python
# Range, estrutura frequentemente usada com for, para gerar uma sequência de números
# range(inicio, fim, passo)

for i in range(1, 10, 2):
    print(i)
```

```python
# Operadores compostos, são usados para realizar operações em uma variável,
# E atribuir o resultado de volta a mesma, +=, *=. -=, /=

soma = 0
for i in range (1,6): #soma os números de 1 a 5
    soma += i
print(soma)
```

```python
# Contagem com passo
for i in range (0, 10, 2):
    print(i)
```

```python
# Contagem reversa
for i in range (10, 0, -1):
    print(i)
```

```python
# Listando pares
pares = []
for i in range(20):
    if i % 2 == 0:
        pares.append(i)
print("Numeros pares:", pares)
```

```python
# While executa um bloco de código enquanto uma condição for verdadeira

contador = 0
while contador < 5:
    print(contador)
    contador +=1
```

```python
# Soma de números
soma = 0
numero = 1
while numero <= 5:
    soma += numero
    numero += 1
    print("A soma é:", soma)
```

```python
# Validando entrada de usuario

numero = -1
while numero < 0:
    numero = int(input("Digite um número positivo: "))
print("Você digitou:", numero)
```

```python
# Loop infinito
while True:
    resposta = input("Digite 'sair' para encerrar: ")
    if resposta == 'sair':
        break
```

```python
# Soma de números pares
soma = 0
numero = 1

while numero <= 100:
        if numero <= 100:
          if numero % 2 == 0:
           soma += numero
        numero += 1
print("A soma dos numeros pares de 1 a 100 é:", soma)
```

```python
# Tabuada
numero = int(input("Digite um número para gerar a tabuada: "))
contador = 1

while contador <= 10:
      print(f"{numero} * {contador} = {numero * contador}")
      contador += 1
```

```python
# Contador de vogais

frase = input("Digite uma frase: ")
frase = frase.lower()
contador = 0
vogais = "aeiou"

for letra in frase:
    if letra in vogais:
        contador += 1
print("Numero de vogais na frase:", contador)
```
