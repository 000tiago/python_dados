Operadores Lógicos

- And
- Or
- Not

```python
print(10 > 2 and 4 < 6)  # True and True  -> True
print(10 > 2 and 4 > 6)  # True and False -> False
print(10 < 2 and 4 < 6)  # False and True -> False
print(10 < 2 and 4 > 6)  # False and False -> False

print(10 > 2 or 4 < 6)   # True or True  -> True
print(10 > 2 or 4 > 6)   # True or False -> True
print(10 < 2 or 4 < 6)   # False or True -> True
print(10 < 2 or 4 > 6)   # False or False -> False

print(not (10 > 2))  # not True  -> False
print(not (10 < 2))  # not False -> True
print(not (10 > 2 or 4 > 6))   # not (True or False)  -> not True  -> False
print(not (10 < 2 and 4 > 6))  # not (False and False) -> not False -> True
```
